**First importing the necessary libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Creating Our data**

```python
Year = [2012, 2014, 2016, 2020, 2021, 2022, 2018]
Profit = [80, 75.8, 74, 65, 99.5, 19, 33.6]
```

**Creating a dataframe from that data**

```python
my_df_year_profit = pd.DataFrame({"Year": Year, "Profit": Profit})
my_df_year_profit
```

**Visualizing with lineplot (matlplot library) the dataframe**

```python
sns.lineplot(x= "Year", y= "Profit", data= my_df_year_profit )
plt.show()
```

**Add hue to add a third variable to the visualization**

**So, let's add a new column call it 'Store' to our previous df and visualize everything in the lineplot**

```python
my_df_year_profit["Store"] = ['A', 'B', 'A', 'B', 'B', 'A', 'A']
my_df_year_profit
```

```python
sns.lineplot(x="Year", y="Profit", data= my_df_year_profit, hue='Store')
```
