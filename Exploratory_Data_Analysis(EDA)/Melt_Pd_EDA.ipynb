**Import the necessary libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Create the data and transform into a dataframe**

```python
data = {
    'Year': [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016],
    'Goals Messi': [41, 60, 59, 91, 45, 58, 52, 59],
    'Assists Messi': [15, 17, 36, 22, 16, 22, 26, 31],
    'Goals Ronaldo': [48, 60, 63, 69, 61, 57, 55, 55],
    'Assists Ronaldo': [6, 18, 18, 13, 15, 20, 17, 16]
}

df = pd.DataFrame(data)
df
```

**Use melt function: melt function in pandas is used to reshape a DataFrame from wide format to long format**

- After applying melt to the DataFrame, you will see that the DataFrame is reshaped into a long format where the 'Goals Messi', 'Assists Messi', 'Goals Ronaldo', and 'Assists Ronaldo' columns are melted into two columns: 'Player' and 'Value'. This transformation makes it easier to analyze and visualize the data based on different player statistics.

Break down to the code:
- *`id_vars`*: This parameter specifies the column(s) that will be retained in the resulting DataFrame without being melted. In this case, 'Year' is specified as the id_vars, so it will remain as a column in the melted DataFrame.
- *`value_vars`*: This parameter specifies the columns that will be melted down. In this case, 'Goals Messi' and 'Goals Ronaldo' are specified as the value_vars, so these columns will be melted into a single column.
- *`var_name`*: This parameter specifies the name of the new column that will store the variable names from the melted columns. In this case, it will be named 'Variable'.
- *`value_name`*: This parameter specifies the name of the new column that will store the values from the melted columns. In this case, it will be named 'Goals'.

```python
melted_df = df.melt(id_vars='Year', value_vars=['Goals Messi', 'Goals Ronaldo'], var_name='Variable', value_name='Goals')
melted_df
```

```python
plot_goals_messi_ronaldo = sns.barplot(x = "Year", y="Goals", hue="Variable", data= melted_df)
plt.show()
```
