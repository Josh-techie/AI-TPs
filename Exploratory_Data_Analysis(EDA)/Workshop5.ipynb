# **Exercice III:**

**K-means clustering**

Tout d'abord, nous allons créer un ensemble de données qui se compose de 150 points générés aléatoirement qui sont grossièrement regroupés en trois régions avec une densité élevée, qui est visualisée via un nuage de points deux dimensionnel.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import seaborn as sns

# create dataset
X, y = make_blobs(
   n_samples=150, n_features=2,
   centers=3, cluster_std=0.5,
   shuffle=True, random_state=0
)

```

Nous définissons le nombre de clusters souhaités à **3**. Nous définissons **n_init = 10** pour exécuter les algorithmes de clustering k-means 10 fois indépendamment avec différents centroïdes aléatoires pour choisir le modèle final comme celui avec le SSE le plus bas.
et finalement le paramètre **max_iter** spécifie le nombre maximum d'itérations pour chaque exécution (ici, 300).

Et Finalement nous prédisons les étiquettes de cluster y_km.

```python
km = KMeans(
    n_clusters=3, init='random',
    n_init=10, max_iter=300
)
y_km = km.fit_predict(X)
```

Maintenant que nous avons prédit les étiquettes de cluster y_km,
visualisez les clusters que notre algorithm k-means a identifiés dans l'ensemble de données avec les centroïdes de cluster. Ceux-ci sont stockés sous l'attribut cluster_centers_ de l'objet KMeans:

**Elbow Method**

La méthode du coude est un outil graphique utile pour estimer le nombre optimal de clusters k pour une tâche donnée. Intuitivement, nous pouvons dire que, si k augmente, la SSE intra-cluster ("distorsion") diminuera puisque les échantillons seront plus proches des centroïdes auxquels ils sont affectés.

En utilisant line plot déterminez le meilleur choix pour le nombre des clusters.

Comme nous pouvons le voir dans le graphique résultant, le coude est situé à k = 3, ce qui prouve que k = 3 est en effet un bon choix pour cet ensemble de données.
